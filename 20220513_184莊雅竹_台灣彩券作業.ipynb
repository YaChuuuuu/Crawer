```python
#取得網頁內容
import requests
from bs4 import BeautifulSoup
url='https://www.taiwanlottery.com.tw/index_new.aspx'
r=requests.get(url)
sp=BeautifulSoup(r.text,'lxml')
```

```python
#BINGO BINGO
datas=sp.find('div',class_='contents_box01')
#print(datas) #開獎區塊
title=datas.find('span',class_='font_black15')
#print(title) #開獎期數
nums=datas.find_all('div',class_='ball_tx ball_yellow')
#print(nums) #開獎號碼

print('BINGO BINGO',title.text,'開獎號碼為：')
for i,num in enumerate(nums):
    print(nums[i].text,end=' ')
```

```python
#雙贏彩
datas=sp.find('div',class_='contents_box06')
#print(datas) #開獎區塊
title=datas.find('span',class_='font_black15')
#print(title) #開獎期數
nums=datas.find_all('div',class_='ball_tx ball_blue')
#print(nums) #開獎號碼
#開獎號碼有24個 其中前面一半是開出順序

print('雙贏彩',title.text,'開獎號碼為：')
print('開出順序：',end='')
for i in range(12):
    print(nums[i].text,end=' ')
print()
print('大小順序：',end='')
for i in range(12,24):
    print(nums[i].text,end=' ')
```

```python
#威力彩
datas=sp.find('div',class_='contents_box02')
#print(datas) #開獎區塊
title=datas.find('span',class_='font_black15')
#print(title) #開獎期數
nums=datas.find_all('div',class_='ball_tx ball_green')
#print(nums) #開獎號碼
specialNum=datas.find('div',class_='ball_red')
#print(specialNum) #第二區號碼
#開獎號碼有12個 其中前面一半是開出順序

print('威力彩',title.text,'開獎號碼為：')
print('開出順序：',end='')
for i in range(6):
    print(nums[i].text,end=' ')
print()
print('大小順序：',end='')
for i in range(6,12):
    print(nums[i].text,end=' ')
print()
print('第二區號碼為：',specialNum.text)
```

```python
#38樂合彩
datas=sp.find_all('div',class_='contents_box02')
#print(datas) #開獎區塊
#print(datas[1]) #38樂合彩
title=datas[1].find('span',class_='font_black15')
#print(title) #開獎期數
nums=datas[1].find_all('div',class_='ball_tx ball_green')
#print(nums) #開獎號碼
#開獎號碼有12個 其中前面一半是開出順序

print('38樂合彩',title.text,'開獎號碼為：')
print('開出順序：',end='')
for i in range(6):
    print(nums[i].text,end=' ')
print()
print('大小順序：',end='')
for i in range(6,12):
    print(nums[i].text,end=' ')
```
